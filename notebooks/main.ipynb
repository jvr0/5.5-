```python
import pandas as pd
```

# Importación

```python
df = pd.read_csv('../data/ElArticoDiscos-inventory-raw.csv')

df.columns=[e.lower().replace(' ', '_') for e in df.columns]
```

```python
df.head()
```

# Transformación

```python
df.drop(['catno', 'listed', 'comments', 'external_id', 'location'], axis = 1, inplace=True)
```

```python
df.price = df.price.apply(lambda x: str(x).replace('.',','))
df.artist = df.artist.apply(lambda x: x.split('(')[0])
df.label = df.label.apply(lambda x: x.split('(')[0])
```

```python
def format_string(string):
    CD = False
    vinyl = False
    single = False

    if 'CD' in string:
        CD = True
    elif '7' in string or 'Single' in string:
        single = True
    elif '12' in string or 'Maxi' in string:
        vinyl = True
    else:
        # En caso de que ninguno de los casos anteriores se cumpla, asignamos un valor a vinyl
        vinyl = True

    # Asegurémonos de devolver siempre tres valores booleanos
    return CD, vinyl, single


df[['CD', 'vinyl', 'single']] = df['format'].apply(lambda x: pd.Series(format_string(x)))
df.drop('format', axis=1, inplace=True)
```

# Exportación

```python
df.to_csv('../data/ElArticoDiscos-inventory-clean.csv', index=False)
```
